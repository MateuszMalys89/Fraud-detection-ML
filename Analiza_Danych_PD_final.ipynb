```python
train_fe = train.copy()
train_fe.drop(columns=cols_to_drop_high_missing, inplace=True)
train_fe.shape
```

Na etapie przygotowania danych dokonano podziału zmiennych wejściowych na numeryczne i kategoryczne na podstawie ich typów danych. Taki podział umożliwia zastosowanie odmiennych strategii przetwarzania, w szczególności w zakresie imputacji braków danych oraz kodowania zmiennych kategorycznych

```python
# zmienne numeryczne
num_cols = train_fe.select_dtypes(include=['int64', 'float64']).columns

# zmienne kategoryczne
cat_cols = train_fe.select_dtypes(include=['object']).columns

#s[rawdzam ilość kolumn w zależności od typu zmiennych
len(num_cols), len(cat_cols)
```

Braki danych w zmiennych numerycznych zostały uzupełnione przy użyciu imputacji medianą. Mediana została obliczona na podstawie zbioru treningowego i następnie wykorzystana do uzupełnienia brakujących wartości, co ogranicza wpływ wartości odstających oraz zapobiega wyciekowi informacji.

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy='median')
train_fe[num_cols] = num_imputer.fit_transform(train_fe[num_cols])
```

Zmienne kategoryczne zostały uzupełnione najczęściej występującą wartością, a następnie zakodowane przy użyciu kodowania one-hot. Po usunięciu cech pomocniczych oraz wylosowaniu podzbioru danych w celu ograniczenia złożoności obliczeniowej, dane zostały poddane standaryzacji z wykorzystaniem metody StandardScaler.

```python
cat_imputer = SimpleImputer(strategy='most_frequent')
train_fe[cat_cols] = cat_imputer.fit_transform(train_fe[cat_cols])
```

# Modele ML - Isolation Forest

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_scaled = scaler.fit_transform(X_sample)
```

```python
train_scaled.shape
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
configs = [
    {"n_estimators": 100, "contamination": 0.035},
    {"n_estimators": 200, "contamination": 0.035},
    {"n_estimators": 500, "contamination": 0.035},

    {"n_estimators": 100, "contamination": 0.02},
    {"n_estimators": 200, "contamination": 0.02},
    {"n_estimators": 500, "contamination": 0.02},

    {"n_estimators": 100, "contamination": 0.045},
    {"n_estimators": 200, "contamination": 0.045},
    {"n_estimators": 500, "contamination": 0.045},
]
```

```python
thresholds = [0.005, 0.01, 0.02, 0.05]
```

```python
results = []

for cfg in configs:
    model = IsolationForest(
        n_estimators=cfg["n_estimators"],
        contamination=cfg["contamination"],
        random_state=42,
        n_jobs=-1
    )

    model.fit(train_scaled)

    scores = -model.score_samples(train_scaled)

    for thr in thresholds:
        thr_value = np.percentile(scores, 100 * (1 - thr))
        y_pred = (scores > thr_value).astype(int)

        results.append({
            "n_estimators": cfg["n_estimators"],
            "contamination": cfg["contamination"],
            "threshold": thr,
            "precision": precision_score(y_sample, y_pred),
            "recall": recall_score(y_sample, y_pred),
            "f1": f1_score(y_sample, y_pred),
            "alerts": y_pred.sum()
        })
```

```python
results_df = pd.DataFrame(results)
results_df
```

```python
results_df[results_df["threshold"] == 0.02] \
    .sort_values("f1", ascending=False)
```

```python
results_df[results_df["threshold"] == 0.01] \
    .sort_values("f1", ascending=False)
```

```python
results_df[results_df["threshold"] == 0.05] \
    .sort_values("f1", ascending=False)
```

```python
results_df[results_df["threshold"] == 0.005] \
    .sort_values("f1", ascending=False)
```

```python
# Wizualizacja wyników dla najlepszego modelu
best_cfg = {"n_estimators": 200, "contamination": 0.035}

iso_forest = IsolationForest(
    **best_cfg,
    random_state=42,
    n_jobs=-1
)
iso_forest.fit(train_scaled)

anomaly_score = -iso_forest.score_samples(train_scaled)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(anomaly_score, bins=50)
plt.title("Rozkład miary anomalii – Isolation Forest")
plt.xlabel("Miara anomalii")
plt.ylabel("Liczba transakcji")
plt.tight_layout()
plt.show()
```

# Autoencoder

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
```

```python
# AUTOENCODER – przygotowanie danych
X_ae = train_scaled
y_ae = y_sample   

# tylko transakcje normalne do uczenia
X_train_ae = X_ae[y_ae == 0]
X_train_ae_small = X_train_ae[:30000]

print(X_ae.shape)
print(X_train_ae_small.shape)
```

```python
input_dim = X_ae.shape[1]

input_layer = Input(shape=(input_dim,))
encoded = Dense(128, activation='relu')(input_layer)
encoded = Dense(64, activation='relu')(encoded)

decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder_1 = Model(inputs=input_layer, outputs=decoded)

autoencoder_1.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse'
)
```

```python
history_1 = autoencoder_1.fit(
    X_train_ae_small,
    X_train_ae_small,
    epochs=20,
    batch_size=64,
    shuffle=True,
    validation_split=0.1,
    verbose=1
)
```

```python
batch_size = 64
reconstruction_error_1 = np.zeros(X_ae.shape[0])

for i in range(0, X_ae.shape[0], batch_size):
    X_batch = X_ae[i:i+batch_size]
    X_batch_pred = autoencoder_1.predict(X_batch, verbose=0)

    batch_error = np.mean(
        np.square(X_batch - X_batch_pred),
        axis=1
    )

    reconstruction_error_1[i:i+batch_size] = batch_error
```

```python
print(reconstruction_error_1.shape)
```

```python
from sklearn.metrics import classification_report

threshold_ae_1 = np.percentile(reconstruction_error_1, 98)
alerts_ae_1 = (reconstruction_error_1 >= threshold_ae_1).astype(int)

print(classification_report(y_ae, alerts_ae_1))
```

```python
input_layer = Input(shape=(input_dim,))

encoded = Dense(256, activation='relu')(input_layer)
encoded = Dense(128, activation='relu')(encoded)
encoded = Dense(64, activation='relu')(encoded)

decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(256, activation='relu')(decoded)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder_2 = Model(input_layer, decoded)

autoencoder_2.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse'
)
```

```python
from sklearn.metrics import classification_report
import numpy as np

def evaluate_thresholds(scores, y_true, thresholds):
    results = {}

    for t in thresholds:
        threshold_value = np.percentile(scores, 100 - t)
        alerts = (scores >= threshold_value).astype(int)

        results[t] = classification_report(
            y_true,
            alerts,
            output_dict=True
        )

    return results
```
