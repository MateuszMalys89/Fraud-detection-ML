```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
```

```python
input_dim = X_scaled.shape[1]

input_layer = Input(shape=(input_dim,))
encoded = Dense(128, activation='relu')(input_layer)
encoded = Dense(64, activation='relu')(encoded)

decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
```

```python
autoencoder.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse'
)
```

```python
X_train_ae = X_scaled[y_sample == 0]
```

```python
X_train_ae_small = X_train_ae[:30000]  # pierwsze 30k

history = autoencoder.fit(
    X_train_ae_small,
    X_train_ae_small,
    epochs=20,
    batch_size=64,
    shuffle=True,
    validation_split=0.1,
    verbose=1
)
```

```python
batch_size = 64
reconstruction_error = np.zeros(X_scaled.shape[0])

for i in range(0, X_scaled.shape[0], batch_size):
    X_batch = X_scaled[i:i+batch_size]
    X_batch_pred = autoencoder.predict(X_batch, verbose=0)

    batch_error = np.mean(
        np.square(X_batch - X_batch_pred),
        axis=1
    )

    reconstruction_error[i:i+batch_size] = batch_error
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(
    reconstruction_error[y_sample == 0],
    bins=50,
    alpha=0.6,
    label='Transakcje prawidłowe'
)

plt.hist(
    reconstruction_error[y_sample == 1],
    bins=50,
    alpha=0.6,
    label='Oszustwa'
)

plt.legend(title="Rodzaj transakcji")

plt.title("Rozkład błędu rekonstrukcji – oszustwa i transakcje prawidłowe (Autoencoder)")
plt.xlabel("Błąd rekonstrukcji")
plt.ylabel("Liczba transakcji")
plt.tight_layout()
plt.show()
```

```python
threshold_ae = np.percentile(reconstruction_error, 98)
alerts_ae = (reconstruction_error >= threshold_ae).astype(int)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_sample, alerts_ae))
```

```python
from sklearn.metrics import classification_report

def evaluate_thresholds(errors, y_true, thresholds):
    results = {}

    for t in thresholds:
        threshold_value = np.percentile(errors, 100 - t)
        alerts = (errors >= threshold_value).astype(int)

        print(f"\n===== ALERTY: top {t}% =====")
        print(classification_report(y_true, alerts))

        results[t] = classification_report(
            y_true,
            alerts,
            output_dict=True
        )

    return results
```

```python
thresholds = [0.5, 1, 2, 5]

ae_results = evaluate_thresholds(
    reconstruction_error,
    y_sample,
    thresholds
)
```

```python
if_results = evaluate_thresholds(
    anomaly_score,
    y_sample,
    [0.5, 1, 2, 5]
)
```
