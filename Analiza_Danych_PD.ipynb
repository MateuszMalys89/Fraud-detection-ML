# Przygotowanie danych do ML

```python
#wyznaczamy kolumny z brakami przekraczającymi próg 90%
high_missing = missing_percent[missing_percent > 0.9]
high_missing
```

```python
#ilość kolumn spełniające wymagania - przeznaczone do usunięcia
len(high_missing)
```

```python
cols_to_drop_high_missing=high_missing.index.tolist()
```

W kolejnym etapie przygotowania danych utworzono kopię zbioru treningowego, na której przeprowadzono selekcję cech. Zmiennym, w których odsetek brakujących wartości przekraczał 90%, przypisano znikomą wartość informacyjną i usunięto je ze zbioru danych. Pozwoliło to na ograniczenie liczby cech oraz uproszczenie dalszego procesu modelowania.

```python
train_fe = train.copy()
train_fe.drop(columns=cols_to_drop_high_missing, inplace=True)
train_fe.shape
```

Na etapie przygotowania danych dokonano podziału zmiennych wejściowych na numeryczne i kategoryczne na podstawie ich typów danych. Taki podział umożliwia zastosowanie odmiennych strategii przetwarzania, w szczególności w zakresie imputacji braków danych oraz kodowania zmiennych kategorycznych

```python
# zmienne numeryczne
num_cols = train_fe.select_dtypes(include=['int64', 'float64']).columns

# zmienne kategoryczne
cat_cols = train_fe.select_dtypes(include=['object']).columns

#s[rawdzam ilość kolumn w zależności od typu zmiennych
len(num_cols), len(cat_cols)
```

Braki danych w zmiennych numerycznych zostały uzupełnione przy użyciu imputacji medianą. Mediana została obliczona na podstawie zbioru treningowego i następnie wykorzystana do uzupełnienia brakujących wartości, co ogranicza wpływ wartości odstających oraz zapobiega wyciekowi informacji.

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy='median')
train_fe[num_cols] = num_imputer.fit_transform(train_fe[num_cols])
```

Zmienne kategoryczne zostały uzupełnione najczęściej występującą wartością, a następnie zakodowane przy użyciu kodowania one-hot. Po usunięciu cech pomocniczych oraz wylosowaniu podzbioru danych w celu ograniczenia złożoności obliczeniowej, dane zostały poddane standaryzacji z wykorzystaniem metody StandardScaler.

```python
cat_imputer = SimpleImputer(strategy='most_frequent')
train_fe[cat_cols] = cat_imputer.fit_transform(train_fe[cat_cols])
```

```python
train_fe = pd.get_dummies(train_fe, columns=cat_cols, drop_first=True)
```

```python
train_fe = train_fe.drop(columns=['AmountRange'])
```

Dla każdego progu oceniono jakość detekcji przy użyciu standardowych metryk klasyfikacyjnych, co pozwoliło na identyfikację kompromisu pomiędzy liczbą generowanych alertów a skutecznością wykrywania fraudów.

# Autoencoder

```python
input_dim = X_scaled.shape[1]

input_layer = Input(shape=(input_dim,))
encoded = Dense(128, activation='relu')(input_layer)
encoded = Dense(64, activation='relu')(encoded)

decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
```

```python
batch_size = 64
reconstruction_error = np.zeros(X_scaled.shape[0])

for i in range(0, X_scaled.shape[0], batch_size):
    X_batch = X_scaled[i:i+batch_size]
    X_batch_pred = autoencoder.predict(X_batch, verbose=0)

    batch_error = np.mean(
        np.square(X_batch - X_batch_pred),
        axis=1
    )

    reconstruction_error[i:i+batch_size] = batch_error
```

```python
from sklearn.metrics import classification_report

def evaluate_thresholds(errors, y_true, thresholds):
    results = {}

    for t in thresholds:
        threshold_value = np.percentile(errors, 100 - t)
        alerts = (errors >= threshold_value).astype(int)

        print(f"\n===== ALERTY: top {t}% =====")
        print(classification_report(y_true, alerts))

        results[t] = classification_report(
            y_true,
            alerts,
            output_dict=True
        )

    return results
```
